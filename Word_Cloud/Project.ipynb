# Creating a Word Cloud

```python
# Importing all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
from wordcloud import STOPWORDS,WordCloud

# Reading the dataset
try:
    data = pd.read_csv("Dataset.csv")
except FileNotFoundError:
    print("Dataset file not found.")
# Cleaning the dataset
text_series = data["review"].dropna().astype(str).head(20000)  # Dropping null values and selecting rows
text_series = (
    text_series
    .str.replace(r'[^A-Za-z\s]', '', regex=True) # Removing special symbols from data and lowering case
    .str.lower()
)
text = " ".join(text_series)  # Merging the output string to an empty string
stopwords = set(STOPWORDS)       # Contains super common words like is,am,are,the
stopwords.update([
    "movie","film","one","will","can","make","story","character"   # adding some more words.
])
text = " ".join([word for word in text.split() if word not in stopwords])  # Removing all the stopwords
# Creating the wordcloud
wordcloud = WordCloud(width=1200, height=600, background_color='black',colormap="cool",max_words=200,contour_width=1,contour_color="white").generate(text)    # Generating the WordCloud
plt.figure(figsize=(12, 6),dpi=350)  # Adjusting the size of figure
plt.imshow(wordcloud, interpolation='bilinear')  # Ploting the graph/Cloud
plt.axis('off')    # Removing the axis from the graph/plot
plt.title("IMDB Movie Reviews Word Cloud")     # Setting up the title
plt.show()     # Display
```
